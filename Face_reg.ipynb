```python
import os                           #for seemeless intraction with os
import warnings
warnings.simplefilter('ignore')     #to ignore warning message

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
%matplotlib inline

from skimage.io import imread,imshow
from skimage.transform import resize       #Image Reading,Transform,Colorchange
from skimage.color import rgb2gray
```

```python
srk= os.listdir("/content/srk")
```

```python
chris=os.listdir("/content/Chris")
```

```python
olsan=os.listdir("/content/olsan")
```

```python
limit=50
Shahruk_image= [None]*limit
j=0

for i in srk:
    if(j<limit):
      Shahruk_image[j]=imread("/content/srk/"+i)
      j+=1  
    else:
        break
```

```python
limit=50
elizabeth_image= [None]*limit
j=0

for i in olsan:
    if(j<limit):
      elizabeth_image[j]=imread("/content/olsan/"+i)
      j+=1  
    else:
        break
```

```python
limit=50
hemsworth_image= [None]*limit
j=0

for i in chris:
    if(j<limit):
      hemsworth_image[j]=imread("/content/Chris/"+i)
      j+=1  
    else:
        break
```

```python
Shahruk_grey=[None]*limit
j=0

for i in srk:
    if(j<limit):
        Shahruk_grey[j]=rgb2gray(Shahruk_image[j])
        j+=1
    else:
        break
```

```python
elizabeth_grey=[None]*limit
j=0

for i in olsan:
    if(j<limit):
        elizabeth_grey[j]=rgb2gray(elizabeth_image[j])
        j+=1
    else:
        break
```

```python
hemsworth_grey=[None]*limit
j=0

for i in chris:
    if(j<limit):
        hemsworth_grey[j]=rgb2gray(hemsworth_image[j])
        j+=1
    else:
        break
```
